```python
import qiskit
import pylatexenc
```

```python
from numpy import pi as pi
```

```python
from qiskit import qasm3 as qasm
```

```python
from qiskit import QuantumCircuit as qc
```

```python
qc = qiskit.QuantumCircuit(5)
for i in range(5):
    qc.h(i)
    qc.rx(pi/4,i)
qc.barrier()
qc.t(4)
qc.p(pi/3, 3)
qc.ry(pi/2, 2)
qc.ry(pi/2, 1)
qc.barrier()
for i in range(4):
    qc.cz(i, i+1)

```

```python
qc.measure_all()
```

```python
qc.draw('mpl', style='iqx')
```

```python
print(qasm.dumps(qc))
```

```python
with open('My_QASM_circuit.qasm', 'w') as f:
	qasm.dump(qc, f)
```
